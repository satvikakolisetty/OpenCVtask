```python
import cv2
import numpy

image1 = cv2.imread('superman.jpeg')
image1 = cv2.resize(image1,(800,500))
cv2.imshow('superman',image1)
if cv2.waitKey() == 13:
    pass
cv2.destroyAllWindows()
image200 = cv2.imread('thor.jpeg')
image200 = cv2.resize(image200,(800,500))
image2 = cv2.imread('thor.jpeg')
image2 = cv2.resize(image2,(800,500))
cv2.imshow('thor',image2)
if cv2.waitKey() == 13:
    pass
cv2.destroyAllWindows()
 
#-------------------------------task4.2---------------------------------


#superman face cropping
crop11 = image1[15:145,350:468]
cv2.imshow('superman',crop11)
if cv2.waitKey() == 13:
    pass
cv2.destroyAllWindows()

#thor face cropping
crop2 = image2[15:145,342:460]
cv2.imshow('thor',crop2)
if cv2.waitKey() == 13:
    pass
cv2.destroyAllWindows()

#thor image 
timage = image2
timage[15:145,342:460] = crop11[:,:]
cv2.imshow('thor with superman face',timage)
if cv2.waitKey() == 13:
    pass
cv2.destroyAllWindows()

#thor face cropping ........
crop2 = image200[15:145,342:460]

#superman image
simage = image1
simage[15:145,350:468] = crop2[:,:]
cv2.imshow('superman with thor face',simage)
if cv2.waitKey() == 13:
    pass
cv2.destroyAllWindows()

#--------------------------task4.3----------------------------------
img1 = cv2.imread('superman.jpeg')
img1 = cv2.resize(img1,(800,500))
img2 = cv2.imread('thor.jpeg')
img2 = cv2.resize(img2,(800,500))
#before cropping.
horizantal1 = numpy.hstack((img1,img2))
cv2.imshow('before cropping',horizantal1)
if cv2.waitKey() == 13:
    pass
cv2.destroyAllWindows()
#after cropping
horizantal2 = numpy.hstack((simage,timage))
cv2.imshow('after cropping',horizantal2)
if cv2.waitKey() == 13:
    pass
cv2.destroyAllWindows()
#joining two horizantal stacks vertically
vertical = numpy.vstack((horizantal1,horizantal2))
cv2.imshow('collage',vertical)
if cv2.waitKey() == 13:
    pass
cv2.destroyAllWindows()
```
